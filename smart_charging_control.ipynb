```python
#import charging duration of the cars (csv file)
import csv

charging_duration = []

with open("charging_duration.csv") as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    
    for row in reader:
        try:
            # Strip leading and trailing whitespace, then convert the string to float and append to the list
            charging_duration.append(int(row[0].strip()))
        except ValueError:
            # Handle the case where conversion is not possible
            print(f"Error converting {row[0]} to int")

# Close the file
csvfile.close()

#create 24 dictionaries for the 24 cars and their charging duration
#each dictionary has 1 key: charging duration
#the dictionaries are stored in a list called cars
cars=[]
for i in range(24):
    cars.append({"charging_duration": charging_duration[i]})
print(cars)


#import the load curve (csv file)
import csv
load_curve = []
with open("load_curve.csv") as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    
    for row in reader:
        try:
            load_curve.append(float(row[0].strip()))
        except ValueError:
            print(f"Error converting {row[0]} to float")

# Close the file
csvfile.close()
print(load_curve)

#create 23 dictionaries for the 23 half hours and their load curve and number of cars charging
#each dictionary has 3 keys: time, load_curve and number_of_cars_charging
#the dictionaries are stored in a list called half_hours
half_hours=[]
for i in range(23):
    half_hours.append({"time": i,"load_curve": load_curve[i],"number_of_cars_charging": 0})
print(half_hours)

#sort the half_hours list by load_curve
half_hours.sort(key=lambda x: x["load_curve"])
print(half_hours)

#for each cars, add 1 to the half_hours with the lowest load_curve until the charging duration is reached
for i in range(24):
    print(cars[i]["charging_duration"])
    for j in range(cars[i]["charging_duration"]):
        half_hours[j]["number_of_cars_charging"]+=1
        half_hours[j]["load_curve"]+=7/24
    half_hours.sort(key=lambda x: x["load_curve"])
print(half_hours)

half_hours.sort(key=lambda x: x["time"])
print(half_hours)

#export half_hours["number_of_cars_charging"] to csv file save in the same folder
import csv
with open('number_of_cars_charging.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    for i in range(23):
        writer.writerow([half_hours[i]["number_of_cars_charging"]])
csvfile.close()
```
